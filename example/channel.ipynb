# channel流の数値計算

```julia
using CFD2d
using Plots
```

```julia
param = CFD2d.Parameter(30, 30, 5.0, 1.0, 1000.0, 10.0, 0.01)
field = CFD2d.Field(param)
condition = CFD2d.Channel()
```

```julia
step = 10
```

```julia
CFD2d.initialize(param, field, condition)
```

```julia
CFD2d.initialize(param, field, condition)
for st in 1:step
    CFD2d.smac(param, field)
    #println("step: $(st)")
end
```
